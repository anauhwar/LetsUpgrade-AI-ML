### Question) Find out the correlation of Attrition with other variables with the help of Mannwhitneyu Test and T Test . For example Attrition with Age, income, employe performance.

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
database = pd.read_csv("C:/Users/vishalshivhare/Desktop/general_data.csv")
```

```python
database.head()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelBinarizer()
database["Attrition"] = label_encoder.fit_transform(database["Attrition"])
database.head()
```

```python
databasea = database.drop_duplicates()
```

```python
database = database.dropna()
```

```python
database.head()
```

```python
# Mannwhitneyu Test
```

```python
from scipy.stats import mannwhitneyu
```

```python
# Attrition VS Disrance From Home
stats,p=mannwhitneyu(databasea.Attrition,database.DistanceFromHome)
print(stats,p)
```

```python
# Attrition VS Total Working Years
stats,p=mannwhitneyu(databasea.Attrition,database.TotalWorkingYears)
print(stats,p)
```

```python
# Attrition VS Year At Company
stats,p=mannwhitneyu(databasea.Attrition,database.YearsAtCompany)
print(stats,p)
```

```python
# Attrition VS Year With Current Manager
stats,p=mannwhitneyu(databasea.Attrition,database.YearsWithCurrManager)
print(stats,p)
```

```python
# Attrition VS Monthly Income
stats,p=mannwhitneyu(databasea.Attrition,database.MonthlyIncome)
print(stats,p)
```

```python
# T Test
```

```python
from scipy.stats import ttest_ind
```

```python
# Attrition VS Disrance from home
stat, p=ttest_ind(database.Attrition,database.DistanceFromHome)
print(stat,p)
```

```python
# Attrition VS Monthly Income
stat, p=ttest_ind(database.Attrition,database.MonthlyIncome)
print(stat,p)
```

```python
# Attrition VS Year At Company
stat, p=ttest_ind(database.Attrition,database.YearsAtCompany)
print(stat,p)
```

```python
# Attrition VS Year With Current Manager
stat, p=ttest_ind(database.Attrition,database.YearsWithCurrManager)
print(stat,p)
```
